```python
# pd.options.mode.chained_assignment = None
# Ensure 'lap' is numeric for proper sorting and calculations
lap_times_2023['lap'] = pd.to_numeric(lap_times_2023['lap'])

# Find the middle and last laps for each race
race_laps = lap_times_2023.groupby('raceId')['lap']
race_middle_laps = race_laps.max().apply(lambda x: x // 2 + 1)  # Middle lap index (1-based)
race_last_laps = race_laps.max()  # Last lap index

# Create dictionaries for quick lookup of middle and last laps
middle_lap_dict = race_middle_laps.to_dict()
last_lap_dict = race_last_laps.to_dict()

# Add middle and last lap flags to the original DataFrame
lap_times_2023['isMiddleLap'] = lap_times_2023.apply(lambda row: row['lap'] == middle_lap_dict[row['raceId']], axis=1)
lap_times_2023['isLastLap'] = lap_times_2023.apply(lambda row: row['lap'] == last_lap_dict[row['raceId']], axis=1)

# Group by raceId and driverId to calculate times for first, middle, and last laps
grouped = lap_times_2023.groupby(['raceId', 'driverId'])

results = []
for (race_id, driver_id), group in grouped:
    group = group.sort_values(by='lap')

    # Extract first lap milliseconds
    first_lap = group.iloc[0]
    print('------------------------')
    print(first_lap)
    print('------------------------')
    # Extract middle and last lap milliseconds based on race-defined flags
    middle_lap = group[group['isMiddleLap']].iloc[0] if not group[group['isMiddleLap']].empty else None
    last_lap = group[group['isLastLap']].iloc[0] if not group[group['isLastLap']].empty else None

    results.append({
        'raceId': race_id,
        'driverId': driver_id,
        'firstLapMilliseconds': first_lap['milliseconds'],
        'middleLapMilliseconds': middle_lap['milliseconds'] if middle_lap is not None else None,
        'lastLapMilliseconds': last_lap['milliseconds'] if last_lap is not None else None
    })

# Convert results to a DataFrame
results_df = pd.DataFrame(results)
display(results_df)

# Copy results to times_normalized
times_normalized = results_df.copy()
```

```python
# Ensure 'lap' is numeric for proper sorting and calculations
lap_times_2023['lap'] = pd.to_numeric(lap_times_2023['lap'])

# Find the last lap and middle lap for each race
race_laps = lap_times_2023.groupby('raceId')['lap']
race_middle_laps = race_laps.max().apply(lambda x: x // 2 + 1)  # Middle lap index (1-based)
race_last_laps = race_laps.max()  # Last lap index

# Create dictionaries for quick lookup of middle and last laps
middle_lap_dict = race_middle_laps.to_dict()
last_lap_dict = race_last_laps.to_dict()

# Add middle and last lap flags to the original DataFrame
lap_times_2023['isMiddleLap'] = lap_times_2023.apply(lambda row: row['lap'] == middle_lap_dict[row['raceId']], axis=1)
lap_times_2023['isLastLap'] = lap_times_2023.apply(lambda row: row['lap'] == last_lap_dict[row['raceId']], axis=1)


# lap_times_2023 = lap_times[lap_times['raceId'].isin(races_2023['raceId'])]

lap_times_2023['milliseconds'] = pd.to_numeric(lap_times_2023['milliseconds'])  # Ensure milliseconds is numeric
fastest_laps = lap_times_2023.groupby(['raceId', 'lap'])['milliseconds'].min().reset_index()
fastest_laps.rename(columns={'milliseconds': 'fastestLapMilliseconds'}, inplace=True)

lap_times_2023 = pd.merge(lap_times_2023, fastest_laps, on=['raceId', 'lap'])

lap_times_2023['lapDelta'] = lap_times_2023['milliseconds'] - lap_times_2023['fastestLapMilliseconds']


# Group by raceId and driverId to calculate deltas for first, middle, and last laps
grouped = lap_times_2023.groupby(['raceId', 'driverId'])

# Create a list to store results
results = []

for (race_id, driver_id), group in grouped:
    # Sort laps to ensure order
    group = group.sort_values(by='lap')

    # Extract first lap delta
    first_lap = group.iloc[0]

    # Extract middle and last lap deltas
    middle_lap = group[group['isMiddleLap']].iloc[0] if not group[group['isMiddleLap']].empty else None
    last_lap = group[group['isLastLap']].iloc[0] if not group[group['isLastLap']].empty else None

    # Append deltas to results
    results.append({
        'raceId': race_id,
        'driverId': driver_id,
        'firstLapDelta': first_lap['lapDelta'],
        'middleLapDelta': middle_lap['lapDelta'] if middle_lap is not None else None,
        'lastLapDelta': last_lap['lapDelta'] if last_lap is not None else None
    })

# Convert results to a DataFrame
results_df = pd.DataFrame(results)
results_df = results_df.merge(drivers_2023[['driverId', 'surname']], how='right', on='driverId')
results_df = results_df.merge(races_2023[['raceId', 'name']], how='right', on='raceId')
display(results_df)
times_delta = results_df.copy()
```

```python
def get_zeros(l): return [0 for _ in range(len(l))]


def get_base_plot(ylabel, size = 10, ylim = 0.01, max_size = 10000):
    fig = plt.figure()
    plt.hlines(1,10000,max_size)
    plt.xlim(0,size)
    plt.ylim(0, ylim)
    plt.xlabel("x")
    plt.ylabel(ylabel)

    whole_numbers = [i for i in range(size)]
    y = get_zeros(whole_numbers)
    plt.plot(whole_numbers,y,'|',ms = 10, c="blue")
```

```python
def calculate_estimator(data, h):
    data_len = len(data)
    x = np.linspace(min(data), max(data), 1000)
    estimator = np.zeros(1000)
    for i in data:
        estimator += np.exp(-((x - i) / h) ** 2 / 2)
    estimator = 1 / ((2 * np.pi) ** 0.5 * data_len * h) * estimator
    return estimator
```

```python
def kmeans(X, k, fixed_seed=True):
    if fixed_seed:
        np.random.seed(0)
    centers = np.random.choice(X, k, replace=False)
    # initial_centers = centers.copy()
    clusters = [{"center": center, "values": []} for center in centers]
    for x in X:
        disntances = [abs(x - cluster["center"]) for cluster in clusters]
        cluster_id = np.argmin(disntances)
        clusters[cluster_id]["values"].append(x)
        clusters[cluster_id]["center"] = np.mean(clusters[cluster_id]["values"])


    print(len(clusters))
    colors = [
    "red", "green", "purple", "yellow", "teal", 
    "blue", "orange", "black", "pink", "brown", 
    "gray", "cyan", "magenta", "lime", "olive", 
    "navy", "maroon", "gold", "silver", "violet", 
    "indigo", "coral", "crimson"
]
    # get_base_plot(ylabel="", size=20, ylim=np.max(X), max_size=50000)
    plt.hist(X,bins=50, density=True, histtype="step", color="black")
    for cluster in clusters:
        plt.plot(cluster["values"],[0 for _ in cluster["values"]],'|',ms = 40, c=colors.pop())
        plt.plot(cluster["center"], 0, 'o', ms=10, c="black")
    plt.title(f"k = {k}")
    plt.show()
    return centers
```

```python
def get_clusters(X,k,fixed_seed=True):
    if fixed_seed:
        np.random.seed(0)
    centers = np.random.choice(X, k, replace=False)
    
    
    kmeans = KMeans(n_clusters=k, init=centers.reshape(-1, 1), n_init=1, max_iter=1)
    kmeans.fit(np.array(X).reshape(-1, 1))
    centers = kmeans.cluster_centers_

    clusters = [[] for _ in range(k)]

    for x in X:
        distances = [np.abs(x - center) for center in centers]
        clusters[np.argmin(distances)].append(x)
        
    return clusters
```
