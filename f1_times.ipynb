```python
from kaggle.api.kaggle_api_extended import KaggleApi

# Set up API
api = KaggleApi()
api.authenticate()

# Download dataset to a specified directory
dataset = 'rohanrao/formula-1-world-championship-1950-2020'
download_path = "./data"
api.dataset_download_files(dataset, path=download_path, unzip=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
races = pd.read_csv(download_path + '/races.csv')
races_2023 = races[races['year'] == 2023]
display(races_2023)
```

```python
lap_times = pd.read_csv(download_path + '/lap_times.csv')
lap_times_2023 = lap_times[lap_times['raceId'].isin(races_2023['raceId'])]
display(lap_times_2023)
```

```python
drivers = pd.read_csv(download_path + '/drivers.csv')
drivers_2023 = drivers[drivers['driverId'].isin(lap_times_2023['driverId'])]
display(drivers_2023)
```

```python
# Ensure 'lap' is numeric for proper sorting and calculations
lap_times_2023['lap'] = pd.to_numeric(lap_times_2023['lap'])

# Find the middle and last laps for each race
race_laps = lap_times_2023.groupby('raceId')['lap']
race_middle_laps = race_laps.max().apply(lambda x: x // 2 + 1)  # Middle lap index (1-based)
race_last_laps = race_laps.max()  # Last lap index

# Create dictionaries for quick lookup of middle and last laps
middle_lap_dict = race_middle_laps.to_dict()
last_lap_dict = race_last_laps.to_dict()

# Add middle and last lap flags to the original DataFrame
lap_times_2023['isMiddleLap'] = lap_times_2023.apply(lambda row: row['lap'] == middle_lap_dict[row['raceId']], axis=1)
lap_times_2023['isLastLap'] = lap_times_2023.apply(lambda row: row['lap'] == last_lap_dict[row['raceId']], axis=1)

# Group by raceId and driverId to calculate times for first, middle, and last laps
grouped = lap_times_2023.groupby(['raceId', 'driverId'])

results = []
for (race_id, driver_id), group in grouped:
    group = group.sort_values(by='lap')

    # Extract first lap milliseconds
    first_lap = group.iloc[0]

    # Extract middle and last lap milliseconds based on race-defined flags
    middle_lap = group[group['isMiddleLap']].iloc[0] if not group[group['isMiddleLap']].empty else None
    last_lap = group[group['isLastLap']].iloc[0] if not group[group['isLastLap']].empty else None

    results.append({
        'raceId': race_id,
        'driverId': driver_id,
        'firstLapMilliseconds': first_lap['milliseconds'],
        'middleLapMilliseconds': middle_lap['milliseconds'] if middle_lap is not None else None,
        'lastLapMilliseconds': last_lap['milliseconds'] if last_lap is not None else None
    })

# Convert results to a DataFrame
results_df = pd.DataFrame(results)
display(results_df)

# Copy results to times_normalized
times_normalized = results_df.copy()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(times_normalized['firstLapMilliseconds'], bins=20, edgecolor='black')
plt.xlabel('First Lap Time')
plt.ylabel('Frequency')
plt.title('Distribution of First Lap Time')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(times_normalized['middleLapMilliseconds'], bins=20, edgecolor='black')
plt.xlabel('Middle Lap Time')
plt.ylabel('Frequency')
plt.title('Distribution of Middle Lap Time')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(times_normalized['lastLapMilliseconds'], bins=100, edgecolor='black')
plt.xlabel('Last Lap Time')
plt.ylabel('Frequency')
plt.title('Distribution of Last Lap Time')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(times_normalized[times_normalized['lastLapMilliseconds'] < 1e6]['lastLapMilliseconds'], bins=100, edgecolor='black')
plt.xlabel('Last Lap Time')
plt.ylabel('Frequency')
plt.title('Distribution of Last Lap Time')
plt.show()
```

```python
# Ensure 'lap' is numeric for proper sorting and calculations
lap_times_2023['lap'] = pd.to_numeric(lap_times_2023['lap'])

# Find the last lap and middle lap for each race
race_laps = lap_times_2023.groupby('raceId')['lap']
race_middle_laps = race_laps.max().apply(lambda x: x // 2 + 1)  # Middle lap index (1-based)
race_last_laps = race_laps.max()  # Last lap index

# Create dictionaries for quick lookup of middle and last laps
middle_lap_dict = race_middle_laps.to_dict()
last_lap_dict = race_last_laps.to_dict()

# Add middle and last lap flags to the original DataFrame
lap_times_2023['isMiddleLap'] = lap_times_2023.apply(lambda row: row['lap'] == middle_lap_dict[row['raceId']], axis=1)
lap_times_2023['isLastLap'] = lap_times_2023.apply(lambda row: row['lap'] == last_lap_dict[row['raceId']], axis=1)

# Group by raceId and driverId to calculate deltas for first, middle, and last laps
grouped = lap_times_2023.groupby(['raceId', 'driverId'])

# Create a list to store results
results = []

for (race_id, driver_id), group in grouped:
    # Sort laps to ensure order
    group = group.sort_values(by='lap')

    # Extract first lap delta
    first_lap = group.iloc[0]

    # Extract middle and last lap deltas
    middle_lap = group[group['isMiddleLap']].iloc[0] if not group[group['isMiddleLap']].empty else None
    last_lap = group[group['isLastLap']].iloc[0] if not group[group['isLastLap']].empty else None

    # Append deltas to results
    results.append({
        'raceId': race_id,
        'driverId': driver_id,
        'firstLapDelta': first_lap['lapDelta'],
        'middleLapDelta': middle_lap['lapDelta'] if middle_lap is not None else None,
        'lastLapDelta': last_lap['lapDelta'] if last_lap is not None else None
    })

# Convert results to a DataFrame
results_df = pd.DataFrame(results)
results_df = results_df.merge(drivers_2023[['driverId', 'surname']], how='right', on='driverId')
results_df = results_df.merge(races_2023[['raceId', 'name']], how='right', on='raceId')
display(results_df)
times_delta = results_df.copy()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 6))
for driverId in times_delta['driverId'].unique():
    driver_laps = times_delta[times_delta['driverId'] == driverId]
    plt.scatter(driver_laps['name'], driver_laps['firstLapDelta'], label=driver_laps['surname'].iloc[0])
    plt.xticks(rotation=90)

plt.xlabel('Race')
plt.ylabel('First Lap Delta (ms)')
plt.title('First Lap Delta to Fastest Lap Across Races')
plt.legend(times_delta['surname'].unique(), title='Driver')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create subplots
fig, axes = plt.subplots(3, 1, figsize=(20, 18), sharex=True)

# List of deltas and titles for subplots
lap_deltas = ['firstLapDelta', 'middleLapDelta', 'lastLapDelta']
titles = ['First Lap Delta to Fastest Lap Across Races', 
          'Middle Lap Delta to Fastest Lap Across Races', 
          'Last Lap Delta to Fastest Lap Across Races']

# Iterate through the deltas and create scatter plots for each
for i, (delta, title) in enumerate(zip(lap_deltas, titles)):
    ax = axes[i]
    for driverId in times_delta['driverId'].unique():
        driver_laps = times_delta[times_delta['driverId'] == driverId]
        ax.scatter(driver_laps['name'], driver_laps[delta], label=driver_laps['surname'].iloc[0])
    ax.set_title(title)
    ax.set_ylabel(f'{delta} (ms)')
    ax.grid(True)
    ax.legend(title='Driver')

# Set common x-label and rotate x-ticks
axes[-1].set_xlabel('Race')
for ax in axes:
    ax.tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```
